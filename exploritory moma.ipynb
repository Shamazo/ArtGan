```python
dates = df['Date']
# dates.select_dtypes(include=[str])
# dates[dates.apply(lambda x: isinstance(x, str))]
counter = 0
def convert_date_to_int(date):
    
    try: 
        return int(date)
    except:
        pass
    
    indate = copy.deepcopy(date)
    if pd.isnull(date):
        return np.nan
    
    split_dates = date.split("–")
    split_dates.extend(date.split("-"))
    dates_lst = []
    for dates in split_dates:
        dates_lst.extend(re.findall('\d{4}', date))
      
    try: 
        return int(dates_lst[0])
    except:
        return np.nan
    
df['Date'] = df['Date'].apply(lambda x: convert_date_to_int(x))
print(len(df))
df = df[df['ThumbnailURL'].notnull()]
print(len(df))
df = df[df['Date'] >= 1850]
df = df[df['Date'] <= 2000]
photo =['Photograph']
df = df[df['Classification'].isin(classes)]
df = df[~df['Artist'].isin(['Kirsten Hawthorne'])]
print(len(df))
```
